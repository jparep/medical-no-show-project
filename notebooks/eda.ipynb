```python
import pandas as pd

# Load data from a CSV file
data = pd.read_csv('../data/noshowappointments.csv')

# Display the first few rows of the dataframe
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Age'] = data['Age'].abs()
data.Age.unique()
```

```python
data["ScheduledDay"] = pd.to_datetime(data["ScheduledDay"], errors='coerce')
data.ScheduledDay.unique()
```

```python

data["AppointmentDay"] = pd.to_datetime(data["AppointmentDay"], errors='coerce')
data.AppointmentDay.unique()
```

```python
data["No_Show"] = data["No-show"]
data = data.drop(columns=["No-show"])
data.info()
```

```python
# Save the clean DataFrame to a CSV file
data.to_csv('../data/processed_data.csv', index=False)

print("Cleaned data saved to 'clean_data.csv'")
```
